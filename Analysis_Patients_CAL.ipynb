```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Patient data analysis 

```python
dfmeta=pd.read_excel('Fistulas barquin2.xlsx')
dfmeta.head()
```

```python
# We change Nan values in the fistula feature with 'no' values.

dfmeta.fistula=dfmeta.fistula.fillna('no')
dfmeta
```

```python
dfmeta[dfmeta.fistula == 'si']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# We transform the features: Diabetes, Obesidad and Cardiopatia in numerical variables 

le.fit(dfmeta[['Diabetes']])
dfmeta['Diabetes'] = le.transform(dfmeta['Diabetes'])
le.fit(dfmeta[['Obesidad']])
dfmeta['Obesidad'] = le.transform(dfmeta['Obesidad'])
le.fit(dfmeta[['Cardiopatia']])
dfmeta['Cardiopatia'] = le.transform(dfmeta['Cardiopatia'])

```

```python
df_numerical = dfmeta.select_dtypes(include = ['float64', 'int64'])
```

```python
df_numerical.drop(['NHC','Unnamed: 7'], axis=1, inplace=True)

```

```python
df_numerical
```

```python
y=dfmeta.fistula
X=df_numerical.fillna(0)
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report, confusion_matrix 
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import f1_score
import itertools
from sklearn.linear_model import LogisticRegression

```

```python
def plot_confusion_matrix(cm, classes, nombre, normalize=False,
                          title='Matriz de confusiÃ³n',
                          cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] 
        print("Matriz de confusiÃ³n normalizada")
    else:
        print('Confusion matrix, without normalization')
    print(cm)
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    cbar=plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black") 
    plt.ylabel('Actual')
    plt.xlabel('Prediction')
    plt.tight_layout()
    plt.savefig(nombre, dpi=300)
```

```python
# First we scale our data

from sklearn.preprocessing import RobustScaler

trans = RobustScaler()
X1 = trans.fit_transform(X)
```

## First Proofs.

We analize a singular model to obtain the best set of parameters

```python
X_train, X_test, y_train, y_test = train_test_split(X1, y,test_size=0.4, random_state=0, stratify=y)
#X_train, X_test, y_train, y_test = train_test_split(X1, y,test_size=0.4, random_state=0, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression

LR1 = LogisticRegression(C=10, solver='liblinear').fit(X_train,y_train)
LR2 = LogisticRegression(C=10,solver='liblinear', penalty='l1').fit(X_train,y_train)
LR3 = LogisticRegression(C=10,solver='liblinear', penalty='l2').fit(X_train,y_train)


```

```python
yhat1 = LR1.predict(X_test)
yhat2 = LR2.predict(X_test)
yhat3 = LR3.predict(X_test)


```

```python
print("Test set Score:{:.2f}".format(LR1.score(X_test,y_test)))
print("Training set Score:{:.2f}".format(LR1.score(X_train,y_train)))
```

```python
print("Test set Score:{:.2f}".format(LR2.score(X_test,y_test)))
print("Training set Score:{:.2f}".format(LR2.score(X_train,y_train)))
```

```python
print("Test set Score:{:.2f}".format(LR3.score(X_test,y_test)))
print("Training set Score:{:.2f}".format(LR3.score(X_train,y_train)))
```

```python
# Calcular la matriz de confusiÃ³n
cnf_matrix = confusion_matrix(y_test, yhat1, labels=['no','si'])
np.set_printoptions(precision=2)
print (classification_report(y_test, yhat1))

# Dibujar la matriz de confusiÃ³n no normalizada
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['No','Yes'],normalize= False,  title='Confusion Matrix',nombre='c10LR.png')
```

```python
# Calcular la matriz de confusiÃ³n
cnf_matrix = confusion_matrix(y_test, yhat2, labels=['no','si'])
np.set_printoptions(precision=2)
print (classification_report(y_test, yhat2))

# Dibujar la matriz de confusiÃ³n no normalizada
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['No','Yes'],normalize= False,  title='Confusion Matrix',nombre='medicalc10l1LR.png')
```

```python
# Calcular la matriz de confusiÃ³n
cnf_matrix = confusion_matrix(y_test, yhat3, labels=['no','si'])
np.set_printoptions(precision=2)
print (classification_report(y_test, yhat3))

# Dibujar la matriz de confusiÃ³n no normalizada
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['No','Yes'],normalize= False,  title='Confusion Matrix',nombre='c10l2LR.png')
```

```python
# get importance
importance0 = pd.Series(LR2.coef_[0], index=X.columns).nlargest(3).iloc[::-1]
importance1=pd.Series(LR2.coef_[0], index=X.columns).nsmallest(3).iloc[::-1]
importance = pd.concat([importance0,importance1], axis=0)


# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```

```python
plt.figure(figsize=(15,10))

importance.plot(kind='barh',color='lightcoral')
ax=plt.gca()
ax.set_xlabel('Weight', fontsize=14)
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
plt.yticks(fontsize=14)
plt.xticks(fontsize=12)
plt.tight_layout()
plt.savefig('LR_metaC10l1.png', dpi=500)
plt.show()
```

### Averaging to 500 experiments

```python
nexp=500
pred_values_patients = np.array([])
importances_rf = {}


for i in range(len(X.columns)):
    importances_rf[X.columns[i]]=0

```

```python

for i in range(nexp):
    
    X_train, X_test, y_train, y_test = train_test_split(X1, y, train_size= 0.4,stratify=y)
    
    "Logistic Regression"
    #lr= LogisticRegression(C=0.1,solver='liblinear', penalty='l2').fit(X_train,y_train)
    
    lr=LogisticRegression(C=100, max_iter=100000,penalty='l1', solver='liblinear').fit(X_train,y_train)
    lr.fit(X_train, y_train)
    
    predictions_patients = lr.predict(X_test)
    pred_values_patients = np.concatenate((pred_values_patients, predictions_patients), axis=None)
    
    for k in range(0, len(X.columns)):
        importances_rf[X.columns[k]]=np.average([lr.coef_[0][k],importances_rf[X.columns[k]]])
    
    
```

```python
importancia = pd.Series(importances_rf)
importancia0 =importancia.nlargest(4).iloc[::-1]
importancia1=importancia.nsmallest(4).iloc[::-1]
importancia10=pd.concat([importancia0,importancia1], axis=0)
```

```python
plt.figure(figsize=(15,10))

importancia10.plot(kind='barh',color='lightcoral')
ax=plt.gca()
ax.set_xlabel('Weights', fontsize=14)
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
plt.yticks(fontsize=14)
plt.xticks(fontsize=12)
plt.tight_layout()
plt.savefig('importance_500_average.png', dpi=500)
plt.show()
```
